# PyTorch Tensor Operations

```python
import torch
```

## Arithmetic Element-wise Operations

```python
a = torch.FloatTensor([[100, 2],
                       [3, 4]])
b = torch.FloatTensor([[2, 2],
                       [3, 3]])
```

```python
a + b
```

```python
a - b
```

```python
a * b
```

```python
a / b
```

```python
a == b
```

```python
a != b
```

```python
a ** b
```

## Inplace Operations

```python
print(a)
print(a.mul(b)) #a*b와 같음
print(a)
print(a.mul_(b)) #언더바 사용해서 a를 바꿔줌
print(a)
```

## Sum, Mean (Dimension Reducing Operations)

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]])
```

```python
print(x.sum())
print(x.mean())
```

```python
print(x.sum(dim=0))
print(x.sum(dim=-1))
```

## Broadcast in Operations (모양이 다른 텐서끼리의 연산)

What we did before,

```python
x = torch.FloatTensor([[1, 2]])
y = torch.FloatTensor([[4, 8]])

print(x.size())
print(y.size())
```

```python
z = x + y
print(z)
print(z.size())
```

**Broadcast feature provides operations between different shape of tensors.**

### Tensor + Scalar

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]])
y = 1

print(x.size())
```

```python
z = x + y
print(z)
print(z.size())
```

### Tensor + Vector

```python
x = torch.FloatTensor([[1, 2],
                       [4, 8]])
y = torch.FloatTensor([3,
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y
print(z)
print(z.size())
```

```python
x = torch.FloatTensor([[[1, 2]]])
y = torch.FloatTensor([3,
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y
print(z)
print(z.size())
```

### Tensor + Tensor

```python
x = torch.FloatTensor([[1, 2]])
y = torch.FloatTensor([[3],
                       [5]])

print(x.size())
print(y.size())
```

```python
z = x + y
print(z)
print(z.size())
```

Note that you need to be careful before using broadcast feature.

### Failure Case

```python
x = torch.FloatTensor([[[1, 2],
                        [4, 8]]])
y = torch.FloatTensor([[1, 2, 3],
                       [4, 5, 6],
                       [7, 8, 9]])

print(x.size()) # (1,2,2)
print(y.size()) #사이즈 자체가 (3,3)이어서 1을 어디에 넣어도 맞출 순 없음

z = x + y 
```
